### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
Faturamento_Loja = round(sum(loja.Preço),2)
Faturamento_Loja2 = round(sum(loja2.Preço),2)
Faturamento_Loja3 = round(sum(loja3.Preço),2)
Faturamento_Loja4 = round(sum(loja4.Preço),2)
```

```python
Faturamento = [Faturamento_Loja , Faturamento_Loja2 , Faturamento_Loja3 , Faturamento_Loja4]
lojas = ["loja", "loja2", "loja3", "loja4"]

plt.bar(x=lojas, height=Faturamento)
plt.ylim(1300000, 1550000)
plt.grid(axis='y')
plt.title("Faturamento entre as 4 lojas")
plt.xlabel("Lojas")
plt.ylabel("Faturamento em Milhão")
```

# 2. Vendas por Categoria


```python
loja.groupby('Categoria do Produto')['Categoria do Produto'].value_counts().sort_values(ascending=False).plot(kind="bar",title="Vendas por Categoria Loja_1")
plt.ylabel("Quantidade de Vendas")
plt.grid(axis='y')
plt.ylim(150,500)
```

```python
loja4.groupby('Categoria do Produto')['Categoria do Produto'].value_counts().sort_values(ascending=False).plot(kind="bar",title="Vendas por Categoria Loja_4")
plt.ylabel("Quantidade de Vendas")
plt.grid(axis='y')
plt.ylim(150,500)
```

# 3. Média de Avaliação das Lojas

```python
Media_Sum_Loja = sum(loja['Avaliação da compra'])
Media_Sum_Loja2 = sum(loja2['Avaliação da compra'])
Media_Sum_Loja3 = sum(loja3['Avaliação da compra'])
Media_Sum_Loja4 = sum(loja4['Avaliação da compra'])

Media_Len_Loja = len(loja['Avaliação da compra'])
Media_Len_Loja2 = len(loja2['Avaliação da compra'])
Media_Len_Loja3 = len(loja3['Avaliação da compra'])
Media_Len_Loja4 = len(loja4['Avaliação da compra'])

Media_Loja = Media_Sum_Loja / Media_Len_Loja
Media_Loja2 = Media_Sum_Loja2 / Media_Len_Loja2
Media_Loja3 = Media_Sum_Loja3 / Media_Len_Loja3
Media_Loja4 = Media_Sum_Loja4 / Media_Len_Loja4
```

```python
Media_Lojas = [round(Media_Loja,2), round(Media_Loja2,2), round(Media_Loja3,2), round(Media_Loja4,2)]
Lojas = ["loja", "loja2", "loja3", "loja4"]

plt.barh(y=Lojas, width=Media_Lojas)
plt.ylabel("Nome das Lojas")
plt.xlabel("Média de Avaliação")
plt.grid(axis='x')
plt.xlim(3.55, 4.4)
```

# 5. Frete Médio por Loja


```python
Frete_Sum_Loja = sum(loja['Frete'])
Frete_Sum_Loja2 = sum(loja2['Frete'])
Frete_Sum_Loja3 = sum(loja3['Frete'])
Frete_Sum_Loja4 = sum(loja4['Frete'])

Frete_Len_Loja = len(loja['Frete'])
Frete_Len_Loja2 = len(loja2['Frete'])
Frete_Len_Loja3 = len(loja3['Frete'])
Frete_Len_Loja4 = len(loja4['Frete'])

Frete_Media = Frete_Sum_Loja / Frete_Len_Loja
Frete_Media2 = Frete_Sum_Loja2 / Frete_Len_Loja2
Frete_Media3 = Frete_Sum_Loja3 / Frete_Len_Loja3
Frete_Media4 = Frete_Sum_Loja4 / Frete_Len_Loja4
```

```python
Medias_Frete = [round(Frete_Media,2), round(Frete_Media2,2), round(Frete_Media3,2), round(Frete_Media4,2)]
Lojas = ["loja", "loja2", "loja3", "loja4"]

plt.bar(x=Lojas, height=Medias_Frete)
plt.ylim(30.0,35.0)
plt.grid(axis='y')
plt.ylabel("Preço Frete (R$)")
```
